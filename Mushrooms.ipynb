```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
data = data.loc[:,data.isna().sum() < 128880]
data.dropna(axis=0, inplace=True)
```

```python
data.set_index("id", inplace=True)
```

```python
data.head()
```

```python
labelencoder = LabelEncoder()
data["class"]=labelencoder.fit_transform(data["class"])
data["cap-shape"]=labelencoder.fit_transform(data["cap-shape"])
data["cap-color"]=labelencoder.fit_transform(data["cap-color"])
data["does-bruise-or-bleed"]=labelencoder.fit_transform(data["does-bruise-or-bleed"])
data["gill-color"]=labelencoder.fit_transform(data["gill-color"])
data["stem-color"]=labelencoder.fit_transform(data["stem-color"])
data["has-ring"]=labelencoder.fit_transform(data["has-ring"])
data["habitat"]=labelencoder.fit_transform(data["habitat"])
data["season"]=labelencoder.fit_transform(data["season"])
```

```python
data.head()
```

```python
X = data.drop(columns = ["class"])
y = data['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
model = GaussianNB()
from sklearn.decomposition import PCA
pca = PCA(n_components=5)  # Reduce to 20 principal components
X_train_reduced = pca.fit_transform(X_train)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
data_test = pd.read_csv("test.csv")
data_test.set_index("id", inplace=True)
data_test.head()
```

```python
data_test = data_test.loc[:, X.columns]
```

```python
data_test["cap-shape"]=labelencoder.fit_transform(data_test["cap-shape"])
data_test["cap-color"]=labelencoder.fit_transform(data_test["cap-color"])
data_test["does-bruise-or-bleed"]=labelencoder.fit_transform(data_test["does-bruise-or-bleed"])
data_test["gill-color"]=labelencoder.fit_transform(data_test["gill-color"])
data_test["stem-color"]=labelencoder.fit_transform(data_test["stem-color"])
data_test["has-ring"]=labelencoder.fit_transform(data_test["has-ring"])
data_test["habitat"]=labelencoder.fit_transform(data_test["habitat"])
data_test["season"]=labelencoder.fit_transform(data_test["season"])
```
